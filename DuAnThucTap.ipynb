## Đưa ra tình trạng hồ sơ bằng mô hình Hồi quy Logistic



```python
# Biến đầu vào
features = [
    'Tuổi',
    'Tình trạng hôn nhân',
    'Số người phụ thuộc',
    'Thu nhập',
    'Điểm CIC',
    'Chi phí hàng tháng',
    'Trình độ học vấn',
    'Thời gian lao động',
    'Vị trí công việc',
    'Thời gian làm việc tại nơi LV hiện tại',
    'Tình trạng nhà ở hiện tại',
    'Hình thức nhận lương'
]

X = df[features]

# Biến đầu ra
y = df['Tình trạng']
```

```python
y_pred = model.predict(X_test)
```

```python
#Giá trị thực tế và giá trị dự đoán
reg_model_diff = pd.DataFrame({'Actual value': y_test, 'Predicted value': y_pred})
reg_model_diff
```

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

## Đưa ra điểm tín dụng và hạn mức tối đa được cấp bằng mô hình XGBoost


```python
# --- BƯỚC 1: CHUẨN BỊ DỮ LIỆU (Giả sử đã qua bước làm sạch ở trên) ---
XG = df.drop(columns=['Khác Hàng', 'Tình trạng', 'Cấp độ', 'Hạn mức tối đa', 'Điểm tín dụng'])
yG = df['Hạn mức tối đa']

XG_train, XG_test, yG_train, yG_test = train_test_split(XG, yG, test_size=0.2, random_state=42)
```

```python
from xgboost import XGBRegressor
# --- ĐỊNH NGHĨA MÔ HÌNH VÀ LƯỚI THAM SỐ ---
xgb = XGBRegressor(random_state=42)

# Khai báo các giá trị bạn muốn dùng thử
param_grid = {
    'n_estimators': [100, 500, 1000],
    'max_depth': [3, 6, 9],
    'learning_rate': [0.01, 0.1, 0.2]}
```

```python
from sklearn.model_selection import GridSearchCV, train_test_split
# --- CHẠY GRIDSEARCHCV ---
# cv=5 nghĩa là chia dữ liệu làm 5 phần để kiểm tra chéo
# scoring='r2' nghĩa là tìm bộ tham số có chỉ số R-squared cao nhất
grid_search = GridSearchCV(
    estimator=xgb,
    param_grid=param_grid,
    cv=5,
    scoring='r2',
    verbose=1, # Hiển thị quá trình chạy
    n_jobs=-1  # Sử dụng tối đa sức mạnh CPU để chạy nhanh hơn
)

grid_search.fit(XG_train, yG_train)
```

```python
# --- KẾT QUẢ ---
print("Bộ tham số tốt nhất tìm được:")
print(grid_search.best_params_)

print(f"\nĐiểm R2 tốt nhất trên tập huấn luyện: {grid_search.best_score_:.4f}")
```

```python
# Sử dụng mô hình tốt nhất để dự đoán
best_model = grid_search.best_estimator_
y_predXG = best_model.predict(XG_test)
```

```python
#Giá trị thực tế và giá trị dự đoán
reg_model_diff = pd.DataFrame({'Actual value': yG_test, 'Predicted value': y_predXG})
reg_model_diff
```

```python
from sklearn.metrics import mean_absolute_error, r2_score
# Đánh giá mô hình
mae = mean_absolute_error(yG_test, y_predXG)
r2 = r2_score(yG_test, y_predXG)

print(f"Mean Absolute Error (MAE): {mae:.2f}")
print(f"R-squared (R2 Score): {r2:.4f}")
```

```python
import pandas as pd

def preprocess_new_customer(data, scaler, feature_order):
    df_new = pd.DataFrame([data])

    # ===== Mapping =====
    marital_map = {'Không': 0, 'Có': 1}
    education_map = {'Khác': 0, 'Cao đẳng': 1, 'Đại học': 2, 'Sau đại học': 3}
    job_map = {'KD tự do': 0, 'Nhân viên': 1, 'Quản lý': 2}
    housing_map = {'Thuê nhà': 0, 'Nhà trả góp': 1, 'Sở hữu nhà ở': 2}
    salary_map = {'Tiền mặt': 0, 'Nhận lương qua NH khác': 1, 'Nhận lương qua Sacombank': 2}

    df_new['Tình trạng hôn nhân'] = df_new['Tình trạng hôn nhân'].map(marital_map)
    df_new['Trình độ học vấn'] = df_new['Trình độ học vấn'].map(education_map)
    df_new['Vị trí công việc'] = df_new['Vị trí công việc'].map(job_map)
    df_new['Tình trạng nhà ở hiện tại'] = df_new['Tình trạng nhà ở hiện tại'].map(housing_map)
    df_new['Hình thức nhận lương'] = df_new['Hình thức nhận lương'].map(salary_map)

    # ===== Scale =====
    num_cols = [
        'Tuổi',
        'Thu nhập',
        'Điểm CIC',
        'Chi phí hàng tháng',
        'Thời gian lao động',
        'Thời gian làm việc tại nơi LV hiện tại'
    ]

    df_new[num_cols] = scaler.transform(df_new[num_cols])

    # ===== Fill NaN =====
    df_new = df_new.fillna(0)

    # ===== Đảm bảo đúng feature =====
    for col in feature_order:
        if col not in df_new.columns:
            df_new[col] = 0

    df_new = df_new[feature_order]

    return df_new
```

```python
def predict_customer(
    data,
    scaler,
    model_trang_thai,
    model_diem,
    model_han_muc,
    feature_order_trang_thai,
    feature_order_diem,
    feature_order_han_muc
):

    # preprocess riêng cho từng model
    X_status = preprocess_new_customer(data, scaler, feature_order_trang_thai)
    X_diem = preprocess_new_customer(data, scaler, feature_order_diem)
    X_hm = preprocess_new_customer(data, scaler, feature_order_han_muc)

    # predict
    status = model_trang_thai.predict(X_status)[0]
    diem = model_diem.predict(X_diem)[0]

    if status == 1:
        han_muc = model_han_muc.predict(X_hm)[0]

        print("✅ Trạng thái: Được xét duyệt")
        print(f"📊 Điểm tín dụng: {diem:.2f}")
        print(f"💰 Hạn mức tối đa: {han_muc:.2f}")

        return {
            "Trạng thái": "Được xét",
            "Điểm": diem,
            "Hạn mức": han_muc
        }

    else:
        print("❌ Hồ sơ bị từ chối")
        print(f"📊 Điểm tín dụng: {diem:.2f}")

        return {
            "Trạng thái": "Từ chối",
            "Điểm": diem
        }
```

```python
feature_order_trang_thai = X_train.columns.tolist()
feature_order_diem = Xđ_train.columns.tolist()
feature_order_han_muc = XG_train.columns.tolist()
```

```python
result = predict_customer(
    data=customer,
    scaler=scaler,
    model_trang_thai=model_trang_thai,
    model_diem=model_diem,
    model_han_muc=model_han_muc,
    feature_order_trang_thai=feature_order_trang_thai,
    feature_order_diem=feature_order_diem,
    feature_order_han_muc=feature_order_han_muc
)

print(result)
```
